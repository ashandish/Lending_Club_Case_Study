```python
#Now , we will convert the data type to numbers wherever applicable , CSV read version 5 and output to version 6
import pandas as pd
df2 = pd.read_csv("C:\\Users\\riyan\\Downloads\\Case Study\\loan\\modified5_loan.csv")
print(df2.info())
print(df2.term.head(5))
print(df2.term.unique())
# Rename Column term to term_months and revol_util to revol_util_percent
df2.rename(columns={'term' : 'term_months'}, inplace=True)
df2.rename(columns={'revol_util' : 'revol_util_percent'}, inplace=True)
print(df2.info())
#Convert term_months datatype object to numeric
df2['term_months'] = df2['term_months'].str.replace('months', '', regex=False)
df2['term_months']=pd.to_numeric(df2['term_months'])
print(df2.term_months.unique())
print(df2.info())
print(df2.int_rate.head(5))
print(df2.int_rate.unique())
#Rename Column int_rate to int_rate_percent and convert to numeric
df2.rename(columns={'int_rate' : 'int_rate_percent'}, inplace=True)
df2['int_rate_percent'] = df2['int_rate_percent'].str.replace('%', '', regex=False)
df2['int_rate_percent']=pd.to_numeric(df2['int_rate_percent'])
print(df2.info())
df2.to_csv('C:\\Users\\riyan\\Downloads\\Case Study\\loan\\modified6_loan.csv', index=False)
```

```python
# Analyze the impact of interest rate over Charged Off loans 
import pandas as pd
import matplotlib.pyplot as plt

df2 = pd.read_csv("C:\\Users\\riyan\\Downloads\\Case Study\\loan\\modified6_loan.csv")
df2_loan_status_Fully_Paid = df2[df2['loan_status'] == 'Fully Paid']
df2_loan_status_charged_off = df2[df2['loan_status'] == 'Charged Off']

# Break interest rate into intervals (0-5, 5-10, etc.)
bins = [0, 5, 10, 15, 20, 25]
labels = ['0-5', '5-10', '10-15', '15-20', '20-25']
df2['int_rate_interval'] = pd.cut(df2['int_rate_percent'], bins=bins, labels=labels, right=False)

# Calculate the percentage of charged-off loans for each dti interval
charged_off_percent_by_int_rate_interval = df2[df2['loan_status'] == 'Charged Off'].groupby('int_rate_interval').size() / df2.groupby('int_rate_interval').size() * 100

# Plot the percentage of charged-off loans for each interest rate interval
charged_off_percent_by_int_rate_interval.plot(kind='bar', color='salmon', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Interest Rate Interval')
plt.ylabel('Percent of Charged Off Loans')
plt.title('Percent of Charged Off Loans by Interest Rate Interval')

# Show the plot
plt.show()

# Conclusion: Likelihood of loan default increases with higher interest rate which can be observed 
# when int rate increases from 5 - 25 
```

```python
#Draw chart illustrating percent of bad loans by number of open Credit Lines 
import pandas as pd
import matplotlib.pyplot as plt

df2 = pd.read_csv("C:\\Users\\riyan\\Downloads\\Case Study\\loan\\modified6_loan.csv")

# Group by 'open_acc' and calculate the percentage of Charged Off loans
charged_off_percent_by_open_acc = df2[df2['loan_status'] == 'Charged Off'].groupby('open_acc').size() / df2.groupby('open_acc').size() * 100

# Set the size of the figure
plt.figure(figsize=(12, 6))
# Create a bar plot
charged_off_percent_by_open_acc.plot(kind='bar', color='blue')

# Add labels and title
plt.xlabel('No. of Open Credit Lines')
plt.ylabel('Percent of Charged Off Loans')
plt.title('Percent of Charged Off Loans by No. Of Open Credit Lines')
plt.xticks(rotation='horizontal')
# Show the plot
plt.show() 
# Conclusion: No. of open credit lines is a strong indicator of loan default
```

```python
#Draw chart illustrating percent of bad loans by the number of 30+ days 
#past-due incidences of delinquency in the borrower's credit file for the past 2 years
import pandas as pd
import matplotlib.pyplot as plt

df2 = pd.read_csv("C:\\Users\\riyan\\Downloads\\Case Study\\loan\\modified6_loan.csv")

# Group by 'delinq_2yrs' and calculate the percentage of Charged Off loans
charged_off_percent_by_delinq_2yrs = df2[df2['loan_status'] == 'Charged Off'].groupby('delinq_2yrs').size() / df2.groupby('delinq_2yrs').size() * 100

# Set the size of the figure
plt.figure(figsize=(12, 6))
# Create a bar plot
charged_off_percent_by_delinq_2yrs.plot(kind='bar', color='blue')

# Add labels and title
plt.xlabel('No. of 30+ days delinquency in last 2 yrs')
plt.ylabel('Percent of Charged Off Loans')
plt.title('Percent of Charged Off Loans by No. of delinquency')
plt.xticks(rotation='horizontal')
# Show the plot
plt.show() 
# Conclusion: delinq_2yrs is a strong indicator of loan default
```
