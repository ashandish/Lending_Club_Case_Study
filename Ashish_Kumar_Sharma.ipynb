```python
#Fill other columns NA values accordingly  , CSV read version 4 and output to version 5
import pandas as pd
df2 = pd.read_csv("C:\\Users\\riyan\\Downloads\\Case Study\\loan\\modified4_loan.csv")
#drop Null emp_length rows 
df2 = df2.dropna(subset= ['emp_length'])
print(df2.shape)
print(df2.columns)
print(df2.info())
print(df2.emp_length.describe())
print(df2.pub_rec_bankruptcies.describe())
print(df2.pub_rec_bankruptcies.unique())
#fill Null pub_rec_bankruptcies with mode value which is 0
df2['pub_rec_bankruptcies']=df2['pub_rec_bankruptcies'].fillna(0)
print(df2.info())
print(df2.revol_util.unique())
#Replace revol_util to number 
df2['revol_util'] = df2['revol_util'].str.replace('%', '', regex=False)
print(df2.revol_util.unique())
df2['revol_util']=pd.to_numeric(df2['revol_util'])
print(df2.revol_util.describe())
#As Null revol_util are relatively small , we will drop them
df2 = df2.dropna(subset= ['revol_util'])
print(df2.info())
df2.to_csv('C:\\Users\\riyan\\Downloads\\Case Study\\loan\\modified5_loan.csv', index=False)
```

```python
#Now , we will convert the data type to numbers wherever applicable , CSV read version 5 and output to version 6
import pandas as pd
df2 = pd.read_csv("C:\\Users\\riyan\\Downloads\\Case Study\\loan\\modified5_loan.csv")
print(df2.info())
print(df2.term.head(5))
print(df2.term.unique())
# Rename Column term to term_months and revol_util to revol_util_percent
df2.rename(columns={'term' : 'term_months'}, inplace=True)
df2.rename(columns={'revol_util' : 'revol_util_percent'}, inplace=True)
print(df2.info())
#Convert term_months datatype object to numeric
df2['term_months'] = df2['term_months'].str.replace('months', '', regex=False)
df2['term_months']=pd.to_numeric(df2['term_months'])
print(df2.term_months.unique())
print(df2.info())
print(df2.int_rate.head(5))
print(df2.int_rate.unique())
#Rename Column int_rate to int_rate_percent and convert to numeric
df2.rename(columns={'int_rate' : 'int_rate_percent'}, inplace=True)
df2['int_rate_percent'] = df2['int_rate_percent'].str.replace('%', '', regex=False)
df2['int_rate_percent']=pd.to_numeric(df2['int_rate_percent'])
print(df2.info())
df2.to_csv('C:\\Users\\riyan\\Downloads\\Case Study\\loan\\modified6_loan.csv', index=False)
```

```python
#Draw chart illustrating percent of bad loans per grade  
import pandas as pd
import matplotlib.pyplot as plt

df2 = pd.read_csv("C:\\Users\\riyan\\Downloads\\Case Study\\loan\\modified6_loan.csv")

# Group by 'grade' and calculate the percentage of Charged Off loans
charged_off_percent_by_grade = df2[df2['loan_status'] == 'Charged Off'].groupby('grade').size() / df2.groupby('grade').size() * 100

# Create a bar plot
charged_off_percent_by_grade.plot(kind='bar', color='salmon')

# Add labels and title
plt.xlabel('Grade')
plt.ylabel('Percent of Charged Off Loans')
plt.title('Percent of Charged Off Loans by Grade')
plt.xticks(rotation='horizontal')
# Show the plot
plt.show()
#Conclusion: Grade is a strong indicator of loan default. Lower Grades have higher chances of loan default
```

```python
#Draw the Distribution of Annual Income for Charged Off Loans and Fully Paid loans . Ignore the loans with Status as Current 
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

df2 = pd.read_csv("C:\\Users\\riyan\\Downloads\\Case Study\\loan\\modified6_loan.csv")
df2_loan_status_Fully_Paid = df2[df2['loan_status'] == 'Fully Paid']
df2_loan_status_charged_off = df2[df2['loan_status'] == 'Charged Off']

pd.set_option('display.float_format', '{:.0f}'.format)
print(df2.annual_inc.describe())
print(df2_loan_status_charged_off.annual_inc.describe())
print(df2_loan_status_Fully_Paid.annual_inc.describe())

plt.figure(figsize=(12, 6))
# Set x-axis tick format to display true figures
plt.gca().get_xaxis().set_major_formatter(FuncFormatter(lambda x, _: '${:,.0f}'.format(x)))

#Set X axis limit to 200000 to ignore outliers
plt.xlim(0, 200000)


# Create a scatter plot with transparency for chrged Off loans
plt.scatter(df2_loan_status_charged_off['annual_inc'], df2_loan_status_charged_off['loan_status'], color='salmon', alpha=0.2 
           , label='Charged Off')

# Create a scatter plot with transparency for Fully Paid loans
plt.scatter(df2_loan_status_Fully_Paid['annual_inc'], df2_loan_status_Fully_Paid['loan_status'], color='blue', alpha=0.2
            , label='Fully Paid')

# Add labels and title
plt.xlabel('Annual Income')
plt.title('Distribution of Annual Income')
plt.legend()

# Show the plot
plt.show()

#Conclusion: Charged Off Loans are more skewed towards income group below 50000

```
